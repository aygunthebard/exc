```python
import pandas as pd
import numpy as np
```

```python
data_g= np.random.normal (10,1.2,1000)
```

```python
data_g_mean = np.mean(data_g)
print (data_g_mean)
```

```python
data_g_sd = np.std(data_g)
print (data_g_sd)
```

```python
import seaborn as sns
```

```python
sns.displot(data=data_g,bins=10,color="b")
```

```python
hist_data, bin_edges = np.histogram(data_g, bins=10)
```

```python
bin_pos = (bin_edges[:-1]+bin_edges[1:])/2
```

```python
bin_count = hist_data
```

```python
bin_dist = np.abs(bin_pos - data_g_mean )
```

```python
hst_df = pd.DataFrame ({
    'bin_pos': bin_pos,
    'bin_count': bin_count,
    'bin_dist': bin_dist
})
print (hst_df)
```

```python
sns.scatterplot(data=hst_df, x = 'bin_pos', y='bin_count', size= 'bin_dist')
```

```python
bin_dist_r = -(np.abs(bin_pos - data_g_mean ))
bin_dist_r
```

```python
hst_df_r = pd.DataFrame ({
    'bin_pos': bin_pos,
    'bin_count': bin_count,
    'bin_dist_r': bin_dist_r
})
print (hst_df)
```

```python
A = sns.scatterplot(data=hst_df_r, x = 'bin_pos', y='bin_count',size ='bin_dist_r')
```
