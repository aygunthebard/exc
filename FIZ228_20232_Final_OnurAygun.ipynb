```python
def gauss(gx , gy , gq):
    gaussian = (1/(np.sqrt(2*np.pi*gq**2))) * np.exp(- ((gx - gy)**2) / (2 * (gq)**2 ))
    return gaussian
```

```python
def lorentz(lx , lx0 , ly):
    lorentzian = (1 / np.pi)*(ly/ ((lx - lx0)**2 + ly**2))
    return lorentzian
```
